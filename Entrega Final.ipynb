1. Identificación de Outliers
Categorías:
Categoría 1 tiene 17,625 outliers con valores extremos alcanzando hasta 504.68 unidades y $12,236.76 en ventas totales.
Estos valores pueden deberse a eventos como promociones o alta demanda.
Regiones:
Las regiones con mayor cantidad de outliers son TOTAL AUTOS AREA 2 y SCANNING MEXICO.
SCANNING MEXICO destaca con el mayor valor de ventas unitarias (504.68 unidades) y ventas totales ($12,236.76).
Segmentos:
BLEACH presenta 7,206 outliers, seguido por LIQUID & GEL con 3,300.
BLEACH también tiene el outlier más alto tanto en ventas unitarias como en ventas totales.
3. Distribución y Patrones
Por Categoría:
Ventas concentradas en un rango bajo, pero con categorías específicas que sobresalen por sus valores máximos.
Por Región:
Variabilidad significativa entre regiones, con picos en SCANNING MEXICO, probablemente reflejando su importancia en el mercado.
Por Segmento:
Segmentos como BAR y BLEACH tienen valores promedio más altos entre los outliers, indicando su relevancia en eventos excepcionales.

```python
import numpy as np
np.random.seed(42)

fechas = pd.date_range(start="2020-01-01", end="2023-12-31", freq="ME")

ventas_vanish = (
    500 + 
    20 * np.arange(len(fechas)) +
    50 * np.sin(2 * np.pi * np.arange(len(fechas)) / 12) +
    np.random.normal(0, 30, len(fechas))
)

ventas_lysol = (
    300 +
    15 * np.arange(len(fechas)) +
    40 * np.cos(2 * np.pi * np.arange(len(fechas)) / 6) +
    np.random.normal(0, 25, len(fechas))
)

data = pd.DataFrame({
    "Fecha": fechas,
    "Vanish": ventas_vanish.round().astype(float),
    "Lysol": ventas_lysol.round().astype(float)
})

data.loc[data.Fecha.dt.month == 12, "Vanish"] = data.loc[data.Fecha.dt.month == 12, "Vanish"] * 1.5
data.loc[data.Fecha.dt.month == 12, "Lysol"] = data.loc[data.Fecha.dt.month == 12, "Lysol"] * 1.3

data["Vanish"] = data["Vanish"].round().astype(int)
data["Lysol"] = data["Lysol"].round().astype(int)

data.to_csv("ventas_historico.csv", index=False)
data = pd.read_csv('ventas_historico.csv', parse_dates=['Fecha'], index_col='Fecha')

data.dropna(inplace=True)
data
```

```python
train = data.iloc[:-12]
test = data.iloc[-12:]
```

```python
from pmdarima import auto_arima

modelo_vanish = auto_arima(train['Vanish'], seasonal=True, m=12, trace=True)
modelo_vanish.fit(train['Vanish'])
```

```python
modelo_lysol = auto_arima(train['Lysol'], seasonal=True, m=12, trace=True)
modelo_lysol.fit(train['Lysol'])
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
import matplotlib.pyplot as plt

pred_vanish = modelo_vanish.predict(n_periods=12)
mae_vanish = mean_absolute_error(test['Vanish'], pred_vanish)
mape_vanish = mean_absolute_percentage_error(test['Vanish'], pred_vanish) * 100

plt.plot(test.index, test['Vanish'], label='Real')
plt.plot(test.index, pred_vanish, label='Predicción')
plt.legend()
plt.show()
```

```python
pred_Lysol = modelo_lysol.predict(n_periods=12)
mae_Lysol = mean_absolute_error(test['Lysol'], pred_Lysol)
mape_Lysol = mean_absolute_percentage_error(test['Lysol'], pred_Lysol) * 100

plt.plot(test.index, test['Lysol'], label='Real')
plt.plot(test.index, pred_Lysol, label='Predicción')
plt.legend()
plt.show()
```
